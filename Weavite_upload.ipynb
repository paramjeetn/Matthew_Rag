```python
import requests
import json
import time  # Import the time module to use the sleep function

# Configure the batch process with a batch size of 1
client.batch.configure(batch_size=50)

with client.batch as batch:
    for i, entry in enumerate(embedded_chunked_data):
        print(f"Importing data object: {i + 1}")
        properties = {
            "content_id": entry['doc_id'],  # Assuming 'doc_id' should map to 'content_id'
            "video_id": entry['video_id'],
            "duration": float(entry['duration']),
            "start": float(entry['start']),
            "title": entry['title'],
            "likes_count": entry['likes_count'],
            "content": entry['content'],
            "guest": entry['guest'],
        }
        batch.add_data_object(
            data_object=properties,
            class_name=collection_name,
            vector=entry['vector']
        )

        if (i + 1) % 50 == 0:
            print("Pausing for 30 seconds to avoid rate limits...")
            batch.flush()
            time.sleep(30)

```
